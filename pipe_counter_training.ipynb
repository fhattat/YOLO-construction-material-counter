## 4) SAHI ile sayım testi

Çalıştır → boru fotoğrafı yükle → adet + işaretli görsel.
(Dosya adında Türkçe karakter/boşluk olmasın.)

```python
import cv2
from collections import Counter
from sahi import AutoDetectionModel
from sahi.predict import get_sliced_prediction
from google.colab import files
from IPython.display import Image as ColabImage, display

BEST = f'{RUNS_DIR}/faz1/weights/best.pt'
det_model = AutoDetectionModel.from_pretrained(
    model_type='ultralytics', model_path=BEST,
    confidence_threshold=0.3, device='cuda:0',
)

def say_ve_ciz(img_path, out_path='sonuc.jpg', slice_px=640, overlap=0.2):
    result = get_sliced_prediction(
        img_path, det_model,
        slice_height=slice_px, slice_width=slice_px,
        overlap_height_ratio=overlap, overlap_width_ratio=overlap,
    )
    preds = result.object_prediction_list
    sayim = Counter(p.category.name for p in preds)
    img = cv2.imread(img_path)
    for p in preds:
        x1, y1, x2, y2 = map(int, p.bbox.to_xyxy())
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
    ozet = '  '.join(f'{k}:{v}' for k, v in sayim.items()) or 'tespit yok'
    cv2.putText(img, ozet, (15, 45), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 0, 255), 3)
    cv2.imwrite(out_path, img)
    return dict(sayim), out_path

up = files.upload()
sayim, sonuc = say_ve_ciz(list(up.keys())[0])
print('SAYIM:', sayim, '| TOPLAM:', sum(sayim.values()))
display(ColabImage(sonuc, width=900))

# Ayar ipuclari:
#   borular cok kucukse       -> slice_px=512
#   kenarda kopuk tespit      -> overlap=0.3
#   fazla/yanlis tespit       -> confidence_threshold=0.45

# Modeli bilgisayara indirmek icin:
# files.download(BEST)
```
